#Representasi Teks



`Bag-of-Words (BoW)` : merepresentasikan teks sebagai sekumpulan kata yang muncul dalam dokumen tanpa memperhatikan urutan atau konteks kata.




````python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Contoh teks
texts = ["Saya suka belajar pemrograman", "Pemrograman Python sangat menarik", "Saya belajar banyak hal baru"]

# Inisialisasi CountVectorizer
vectorizer = CountVectorizer()

# Melatih model BoW
X = vectorizer.fit_transform(texts)

features = vectorizer.get_feature_names_out()

# Mengonversi BoW menjadi DataFrame
bowdf = pd.DataFrame(X.toarray(), columns=features)
bowdf
````

`TF-IDF` mempertimbangkan frekuensi kata dalam dokumen (TF) dan juga seberapa jarang kata tersebut muncul dalam seluruh corpus (IDF).

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# Inisialisasi TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Melatih model TF-IDF
X_tfidf = tfidf_vectorizer.fit_transform(texts)

features = vectorizer.get_feature_names_out()

# Mengonversi BoW menjadi DataFrame
tf_df = pd.DataFrame(X_tfidf.toarray(), columns=features)
tf_df
````

````python
import pandas as pd

df= pd.read_csv('/content/GojekAppReviewV4.0.0-V4.9.3_Cleaned.csv')
df.head()
````

````python
# Mengambil hanya kolom 'review' dan 'rating'
df = df[['content', 'score']]
df
````

````python
df.isnull().sum()
````

````python
pd.reset_option('display.max_colwidth')
df = df.dropna()
df = df.drop_duplicates()
df
````

````python
import numpy as np
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
````

````python
# Fungsi untuk menghapus tanda baca
def clean_text(text):
    text = text.lower()  # Konversi ke huruf kecil
    text = re.sub(r'[^\w\s]', '', text)  # Hapus tanda baca
    return text

df['content']=df['content'].apply(clean_text)
df.head()
````

````python
def tokenize_text(text):
    return word_tokenize(text)

# Terapkan fungsi tokenisasi pada setiap elemen di kolom 'content'
df['content'] = df['content'].apply(tokenize_text)
df
````

````python
sw = stopwords.words('indonesian')
#print(sw)
# Fungsi untuk menghapus stopwords dari token-token kata


# def remove_stopwords(tokens):
#     filtered_tokens = [word for word in tokens if word.lower() not in sw]
#     return filtered_tokens


def remove_stopwords(tokens):
  filtered_tokens = []
  for word in tokens:
    if word.lower() not in sw:
      filtered_tokens.append(word)
  return filtered_tokens

cekStopword = ["saya,","adalah","manusia"]
print(remove_stopwords(cekStopword))

````

````python
sw = stopwords.words('indonesian')
# Fungsi untuk menghapus stopwords dari token-token kata
def remove_stopwords(tokens):
    filtered_tokens = [word for word in tokens if word.lower() not in sw]
    return filtered_tokens

# Menghapus stopwords dari kolom 'Tokenized Text'
df['content'] = df['content'].apply(remove_stopwords)
df
````

````python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat stemmer dari Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(words):
    return [stemmer.stem(word) for word in words]
````

````python
# Menerapkan fungsi stemming pada kolom 'content'
# Menerapkan stemming hanya pada 5 baris pertama
df.loc[:4, 'content'] = df.loc[:4, 'content'].apply(stemming)
df
````
